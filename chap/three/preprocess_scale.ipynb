```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

import mglearn
```

```python
X, _ = make_blobs(n_samples=50, centers=5, random_state=4, cluster_std=2)

X_train, X_test = train_test_split(X, random_state=5, test_size=0.1)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(13, 4))
axes[0].scatter(X_train[:, 0], X_train[:, 1],
                c=mglearn.cm2(0), label="Training set", s=60)
axes[0].scatter(X_test[:, 0], X_test[:, 1],
                c=mglearn.cm2(1), marker="^", label="Test set", s=60)
axes[0].legend(loc="upper left")
axes[0].set_title("Original Data")


# MinMaxSCalderでデータをスケール変換
scaler = MinMaxScaler()
scaler.fit(X_train)

# X_train, X_testをX_trainに基づくスケール変換を行う
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

axes[1].scatter(X_train_scaled[:, 0], X_train_scaled[:, 1],
                c=mglearn.cm2(0), label="Training set", s=60)
axes[1].scatter(X_test_scaled[:, 0], X_test_scaled[:, 1],
                c=mglearn.cm2(1), marker="^", label="Test set", s=60)


# X_train, X_testとそれぞれのスケールで変換(ダメ)
test_scaler = MinMaxScaler()
test_scaler.fit(X_test)
X_test_scaled_badly = test_scaler.transform(X_test)

axes[2].scatter(X_train_scaled[:, 0], X_train_scaled[:, 1],
                c=mglearn.cm2(0), label="Training set", s=60)
axes[2].scatter(X_test_scaled_badly[:, 0], X_test_scaled_badly[:, 1],
                c=mglearn.cm2(1), marker="^", label="test set", s=60)
axes[2].set_title("Improperly Scaled Data")

for ax in axes:
    ax.set_xlabel("Feature 0")
    ax.set_ylabel("Feature 1")
     
```

```python
from sklearn.svm import SVC
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer["data"], cancer["target"],
                                                    random_state=0)

svm = SVC(C=100, gamma="scale")
svm.fit(X_train, y_train)
print("Test set accuracy", svm.score(X_test, y_test))  # 精度が高いのはsklearnにおけるgammaのdefault値が変更されたため
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

svm.fit(X_train_scaled, y_train)
print("Scaled Test set accuracy", svm.score(X_test_scaled, y_test))
```
